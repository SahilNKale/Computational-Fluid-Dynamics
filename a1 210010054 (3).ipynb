```python
%matplotlib inline
from math import*
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

## Assignment 1

## Q.1

```python
def vel(r,t):
    x,y=r
    if np.sqrt(x*x+y*y)>1:
        return np.array([((x*x-y*y-1)*(x*x-y*y) + 4*(x**2)*(y**2))/((x*x+y*y)**2),(-2*x*y)/((x*x+y*y)**2)])
    else:
        return np.array([0,0])
```

## Pathlines & Streamlines

```python
for j in range(-5,5):
    IC=np.array([-2,j])
    t=np.linspace(0,6,100)
    Sx,Sy=odeint(vel,IC,t).T
    plt.plot(Sx,Sy)
plt.grid()
plt.legend(['Pathlines & Streamlines'],loc='upper right')
plt.xlabel('X')
plt.ylabel('Y')
```

## Velicty Field

```python
x,y=np.mgrid[-2:2:15j,-2:2:15j]
```

```python
V=np.zeros_like(x)
U=np.zeros_like(x)
for i in range(len(x)):
    for j in range(len(y)):
        U[i][j],V[i][j]=vel([x[i][j],y[i][j]],1)
plt.quiver(x,y,U,V,color='r')
```

## Streaklines

```python
for j in range(-5,5):
    IC=np.array([-2.0,j])
    dt=0.1
    n=100
    T=n*dt
    Sx1,Sy1=[],[]
    for i in range(n):
            start = i*dt
            res = odeint(vel, IC, [start, T])
            Sx1.append(res[1,0])
            Sy1.append(res[1,1])
    plt.plot(Sx1, Sy1)
plt.grid()
plt.legend(['Streakline'],loc='upper right')
plt.xlabel('X')
plt.ylabel('Y')
```

## Q.2

```python
def vel2(r,t):
    x,y=r
    return np.array([-((x+t)**2-y**2)/(((x+t)**2+y**2)**2),-2*(x+t)*y/((x+t)**2+y**2)**2])
```

## Streamlines

#### At T=0s

```python
def velt1(r,t=1):
    return vel2(r,0)
for i in range(10):
    IC=np.array([-2,-2+0.1*i])
    t=np.linspace(0,5,100)
    SsX1,SsX2=odeint(velt1,IC,t).T
    plt.plot(SsX1,SsX2)
plt.grid()
plt.axis('equal')
```
